```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("nishaanamin/march-madness-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import mean_squared_error
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
df=pd.read_csv(path+"/TeamRankings.csv")
df.head()
```

```python
Ark_data = df[df['TEAM']=='Arkansas']
```

```python
Ark_data.to_csv('Ark_data.csv', index =False)
```

```python
tmp=Ark_data.drop(Ark_data.index[0])
Ark_data_drop = tmp.reset_index(drop=True)
Ark_data_drop.head()
X_train = Ark_data_drop[['SEED','V 1-25 WINS','V 1-25 LOSS','V 26-50 WINS','V 26-50 LOSS','V 51-100 WINS','V 51-100 LOSS']]
X_test = Ark_data.iloc[[0]][['SEED','V 1-25 WINS','V 1-25 LOSS','V 26-50 WINS','V 26-50 LOSS','V 51-100 WINS','V 51-100 LOSS']]
Y_target = Ark_data_drop['ROUND']
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_target)
```

```python
prediction = reg.predict(X_test)
print(prediction)
```

```python
x, xtest, ytrain, ytest = train_test_split(X_train, Y_target, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(x.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1))
```

```python
model.compile(loss='mse', optimizer=Adam(learning_rate=0.01), metrics=['accuracy'])
```

```python
model.fit(x, ytrain, epochs=200, batch_size=32, validation_data=(xtest, ytest))
```

```python
yhat = model.predict(X_test)

```
